    + Batch learning: (batch): gather a shitload of initial (starting) data, and **whack** that to the ML model for it to learn.

**4. Choose a performance measure**.

**Definitions:**
- Performance measure: A unit/thing that checks **how good/bad a ML model performs.** It's a speedometer of the Machine Learning world.

**Measures/Formula**

Image: RMSE formula (nice for measuring how good regression tasks)

**Example problem**
We are creating a Model that predicts scores of National Exam contestants within Ho Chi Minh City. We have:
- 2000 students participating in the exam.
- Each student has these following info:
  + Number of study hours
  + Number of practice tests
In this case, we have
- m = number of instances = number of students (2000)
- x(i) = a **vector** of the **feature values**
  + Vector: a thing that measures how large/small + direction in space. In this case, it can be understood as a **group of things** belonging to a student.
  + Feature values: Things related to each student (**study hours** + **practice tests**)
  + Example: Student number 2 has the following info:
    + Study hours = 500 hours
    + Practice tests taken = 500 tests
    + x(2) = (500, 500)
- X = a matrix of x(i).
  + Matrix: a **2-dimensional (rows + columns) vector**. Vector (called array in normal programming) is a **series of stuffs** placed next to each other.


  + Simply put, to create X, we stack many x(i) vectors on top of each other. After that, we have:
    + Rows = each x(i) vector (each feature group of student)


- h: Prediction function (a.k.a our main character).
  +  h is an abbreviation for **hypothesis**.
  +  How it works:
    +  Pick an instance (student) and its vector (features of a student - study hours + practice tests)
    +  Feed the vector (features of a student) to the prediction function
